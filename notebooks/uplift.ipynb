```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
df = pd.read_parquet("data/outputs/exp_alloc_demo.parquet")
crosstab = df.groupby(["segment_kmeans","CLV_tier","assign"]).size().reset_index(name="count")
plt.figure(figsize=(8,4))
sns.barplot(data=crosstab, x="CLV_tier", y="count", hue="assign")
plt.title("Treatment vs Control count by CLV_tier (all segments)")
plt.tight_layout(); plt.show()
```
